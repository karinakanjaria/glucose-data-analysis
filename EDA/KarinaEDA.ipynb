```python
# Time series data
a = np.random.randn(10000)
b = np.random.randn(10000)
```

```python
# Zero Mean Cumulative Sum https://variation.com/wp-content/change-point_analyzer_help/hs108.htm
a = fu.toAggregated(a)
b = fu.toAggregated(b)
```

```python
pymfdfa = fathon.MFDFA(a)
```

```python
winSizes = fu.linRangeByStep(10, 2000)
qs = np.arange(-3, 4, 0.1)
revSeg = True
polOrd = 1
```

```python
%%time
n, F = pymfdfa.computeFlucVec(winSizes, qs, revSeg=revSeg, polOrd=polOrd)
```

```python
plt.plot(np.log(n), np.log(F[0, :]), 'ro')
plt.plot(np.log(n), list_H_intercept[0]+list_H[0]*np.log(n), 'k-', label='h_{:.1f} = {:.2f}'.format(qs[0], list_H[0]))
half_idx = int(len(qs)/2)
plt.plot(np.log(n), np.log(F[half_idx, :]), 'co')
plt.plot(np.log(n), list_H_intercept[half_idx]+list_H[half_idx]*np.log(n),
         'k-', label='h_{:.1f} = {:.2f}'.format(qs[half_idx], list_H[half_idx]))
plt.plot(np.log(n), np.log(F[-1, :]), 'yo')
plt.plot(np.log(n), list_H_intercept[-1]+list_H[-1]*np.log(n), 'k-',
         label='h_{:.1f} = {:.2f}'.format(qs[-1], list_H[-1]))
plt.xlabel('ln(n)', fontsize=14)
plt.ylabel('ln(F(n))', fontsize=14)
plt.title('MFDFA', fontsize=14)
plt.legend(loc=0, fontsize=14)
```

```python
plt.plot(qs, list_H, 'ro-')
plt.xlabel('q', fontsize=14)
plt.ylabel('h(q)', fontsize=14)
plt.title('h(q)', fontsize=14)
```

## Wrapped Function

```python
def do_mfdfa(data: list, winSizes: list, qList: list, revSeg: bool, polOrder: int):
    # zero-mean cumulative sum of data
    data_cs = fu.toAggregated(data)
    
    # init mfdfa object
    pymfdfa = fathon.MFDFA(data_cs)
    
    # compute fluctuation function and generalized Hurst exponents
    n, F = pymfdfa.computeFlucVec(winSizes, qList, revSeg, polOrder)
    list_H, list_H_intercept = pymfdfa.fitFlucVec()

    #compute mass exponents
    tau = pymfdfa.computeMassExponents()

    #compute multifractal spectrum
    alpha, mfSpect = pymfdfa.computeMultifractalSpectrum()
    
    return n, F, list_H, list_H_intercept, tau, alpha, mfSpect
```

```python
data = df2.Value.to_numpy()
wins = fu.linRangeByStep(10, 500) # 30 mins to 1/2 year for 
qList = np.arange(1, 400)
revSeg = True
polOrder = 1
```

```python
n, F, list_H, list_H_intercept, tau, alpha, mfSpect = do_mfdfa(data, wins, qList, revSeg, polOrder)
# tau vs window size
# y = list H [x values] + list H intercept
# metric: highest q order point - lowest q order point = difference in the system
# List H slope
```
